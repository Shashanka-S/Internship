1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#calling webdriver
browser = webdriver.Chrome() 

#getting into website
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
browser.get(url) 
time.sleep(2)

#scraping names data
name=[]
try:
    names=browser.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    name = [i.text for i in names]
#Handle NoSuchElementException
except NoSuchElementException:
    name.append('Not available')
#Handle StaleElementReferenceException
except StaleElementReferenceException:
    name.append('Not available')
    
#scraping rank data
rank=[]
try:
    ranks=browser.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    rank = [i.text for i in ranks]
#Handle NoSuchElementException
except NoSuchElementException:
    name.append('Not available')
#Handle StaleElementReferenceException
except StaleElementReferenceException:
    name.append('Not available')

#scraping artist/uploader of the video
artist=[]
try:
    artists=browser.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    artist = [i.text for i in artists]
#Handle NoSuchElementException
except NoSuchElementException:
    artist.append('Not available')
#Handle StaleElementReferenceException
except StaleElementReferenceException:
    artist.append('Not available')
     
#scraping views count data
view=[]
try:
    views=browser.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    view = [i.text for i in views]
#Handle NoSuchElementException
except NoSuchElementException:
    view.append('Not available')
#Handle StaleElementReferenceException
except StaleElementReferenceException:
    view.append('Not available')
      
#scraping upload date data
date=[]
try:
    dates=browser.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    date = [i.text for i in dates]
#Handle NoSuchElementException
except NoSuchElementException:
    date.append('Not available')
#Handle StaleElementReferenceException
except StaleElementReferenceException:
    date.append('Not available')
      
#creating data frame
df=pd.DataFrame({"Rank":rank, "Name":name,"Artist":artist,"Upload Date":date,"Views (billion)":view})
#Displaying data frame
df
```

2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
#calling webdriver
browser = webdriver.Chrome() 

#getting into website
url = 'https://www.bcci.tv/'
browser.get(url) 
time.sleep(2)

#Getting to international fixtures page
i_f = browser.find_element_by_xpath("//li[@class='nav-item']")
try:
    i_f.click()
except ElementNotInteractableException:
    browser.get(i_f.get_attribute('href'))  
time.sleep(2)

#scraping match time data
Time = []
times=browser.find_elements_by_xpath("//h5[@class='text-right ng-binding']")
for i in times:
    if i.text is None :
        Time.append("Not available") 
    else:
        Time.append(i.text)

#Initializing empty lists
title=[] 
series=[]
place=[]
date=[]
Time = []

#Scraping series title data
seriess=browser.find_elements_by_xpath("//div[@class='fixture-card-top']/h5[2]/span")
series = [i.text for i in seriess]
#Scraping match date data
dates=browser.find_elements_by_xpath("//div[@class='match-card-left match-schedule']/h5[1]")
date = [i.text for i in dates]
#Scraping match place data
places=browser.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")
place = [i.text for i in places]
#Scraping match title data
titles=browser.find_elements_by_xpath("//div[@class='fixture-card-mid d-flex align-items-center justify-content-between']")
title = [i.text.replace('\n',' ') for i in titles]
#Scraping match time data
Times=browser.find_elements_by_xpath("//h5[@class='text-right ng-binding']")
Time = [i.text for i in Times]

#creating data frame
df=pd.DataFrame({"Match":title,"Series":series,"Place":place,"Date":date,"Time":Time})
#Displaying data frame
df
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
ASSIGNMENT
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Calling the browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(1)

# Getting into the website
url = "https://github.com/"
driver.get(url)
time.sleep(1)

# clicking on option button
drop=driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[3]/button")
drop.click()
time.sleep(1)

# clicking on trending button
try:
    tr=driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
    tr.click()
#Handling NoSuchElementException
except NoSuchElementException:
    driver.get(url+'trending')
time.sleep(1)

#Initializing empty lists
url=[]
contributor=[]
language=[]
title=[]
description=[]

#scraping repositories urls data
urls=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
url = [i.get_attribute('href') for i in urls]
    
#scraping repositories title data
titles=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
title = [i.text for i in titles]
    
#scraping repositories description data
descriptions=driver.find_elements_by_xpath("//article[@class='Box-row']/p")
description = [i.text for i in descriptions]

for i in url:
    driver.get(i)
    time.sleep(1)
    #scraping contributors count data
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributor.append(count.text)
    #Handle NoSuchElementException
    except NoSuchElementException:
        contributor.append('Not available')
    
    #scraping languages data
    lan = []
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:    
        lan = [i.text for i in languages]
    else:
        lan.append('Not available')
    language.append(lan) 
    
#creating data frame
df=pd.DataFrame({'Title':title,'Description':description,'Contributors count':contributor,'Language used':language})
#Displaying data frame
df
```
