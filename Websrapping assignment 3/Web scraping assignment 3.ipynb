4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver = webdriver.Chrome("chromedriver.exe") #calling webdriver
driver.get('https://www.flipkart.com')  #getting into website  
driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]').click() #closing login pop-up
driver.find_element_by_xpath("//input[@name='q']").send_keys('Oneplus Nord') #Entering required input in product search field
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click() #clicking search button
driver.maximize_window() #Maximizing the window

br = []
nm = []
cr = []
ra = []
ro = []
pc = []
sc = []
ds = []
bc = []
pr = []
ur = []

title_tags = driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in title_tags: #Scraping brand, name, color data
    br.append(i.text.split()[0])
    nm.append(i.text.split('(')[0])
    cr.append(i.text.split('(')[1].replace(')','').split(',')[0])

storage_tags = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[1]")
for i in storage_tags: #Scraping storage RAM, ROM data
    ra.append(i.text.split('|')[0])
    ro.append(i.text.split('|')[1])

camera_tags = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[3]")
for i in camera_tags: #Scraping primary and secondary camera data
    pc.append(i.text.split('|')[0])
    try:
        sc.append(i.text.split('|')[1])
    except IndexError:
        sc.append('-')

disp_tags = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[2]")
for i in disp_tags: #Scraping display data
    ds.append(i.text)
    
bat_tags = driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[4]")
for i in bat_tags: #Scraping battery data
    bc.append(i.text)
    
price_tags = driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for i in price_tags: #Scraping price data
    pr.append(i.text)
    
url_tags = driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in url_tags: #Scraping url data
    ur.append(i.get_attribute('href'))
    
data = pd.DataFrame({'Brand':br, 'Product name':nm, 'Colour':cr, 'RAM':rm, 'ROM':rm, 
                     'Primary camer':pc, 'Secondary camera':sc, 'Display size':ds, 
                     'Battery':bc, 'Price':pr, 'URL':ur }) #saving as dataframe
data.to_csv('Scrapped.csv') #saving to a csv file
data #displaying
```

6. Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) from trak.in.

```python
driver = webdriver.Chrome("chromedriver.exe") #calling webdriver
driver.get('https://trak.in/')  #getting into website
driver.maximize_window() #Maximizing the window
driver.find_element_by_xpath("//li[@id='menu-item-51510']").click() #clicking funding deals
sl = []
dt = []
st = []
ind = []
su = []
ci = []
iv = []
it = []
am = []
mar = '//table[@id="tablepress-56"]/tbody/tr/td' #for march 2021
feb = '//table[@id="tablepress-55"]/tbody/tr/td' #for feb 2021
jan = '//table[@id="tablepress-54"]/tbody/tr/td' #for jan 2021
months = [jan, feb, mar]
col = [sl, dt,st, ind,su,ci,iv,it,am]
for j in months: #Month wise for second quarter
    for q in range(1,10): #scrapping each column wise data
        path = driver.find_elements_by_xpath(j+'['+str(q)+']')
        for i in path: #Scrapping each row wise data
             col[q-1].append(i.text)   
time.sleep(2)
driver.close() #Closing the browser
second_quarter = pd.DataFrame({'Sl no':sl, 'Date':dt, 'Startup name':st, 'Industry/vertical':ind, 
                               'Sub vertical':su, 'City':ci, 'Investor name':iv, 'Investment type':it, 
                              'Amount':am }) #creating dataframe of scraped data
second_quarter #displaying scrapped data
```
