6. Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) from trak.in.

```python
driver = webdriver.Chrome("chromedriver.exe") #calling webdriver
driver.get('https://trak.in/')  #getting into website
driver.maximize_window() #Maximizing the window
driver.find_element_by_xpath("//li[@id='menu-item-51510']").click() #clicking funding deals
sl = []
dt = []
st = []
ind = []
su = []
ci = []
iv = []
it = []
am = []
mar = '//table[@id="tablepress-56"]/tbody/tr/td' #for march 2021
feb = '//table[@id="tablepress-55"]/tbody/tr/td' #for feb 2021
jan = '//table[@id="tablepress-54"]/tbody/tr/td' #for jan 2021
months = [jan, feb, mar]
col = [sl, dt,st, ind,su,ci,iv,it,am]
for j in months: #Month wise for second quarter
    for q in range(1,10): #scrapping each column wise data
        path = driver.find_elements_by_xpath(j+'['+str(q)+']')
        for i in path: #Scrapping each row wise data
             col[q-1].append(i.text)   
time.sleep(2)
driver.close() #Closing the browser
second_quarter = pd.DataFrame({'Sl no':sl, 'Date':dt, 'Startup name':st, 'Industry/vertical':ind, 
                               'Sub vertical':su, 'City':ci, 'Investor name':iv, 'Investment type':it, 
                              'Amount':am }) #creating dataframe of scraped data
second_quarter #displaying scrapped data
```

7. Write a program to scrap all the available details of best gaming laptops from digit.in

```python
driver = webdriver.Chrome("chromedriver.exe") #calling webdriver
driver.get('https://digit.in/')  #getting into website
driver.maximize_window() #Maximizing the window
driver.find_element_by_xpath("//div[@class='menu']/ul/li[4]").click() #clicking top 10
driver.find_element_by_xpath("//div[@class='categoty_list']/button[2]").click() #clicking on laptops
driver.find_element_by_xpath("//*[text()='Best Gaming Laptops in India']").click() #clicking on best gaming laptops
nm = []
pr = []
ds = []
os = []
me = []
gr = []
bd = []
ur = []
NM = driver.find_elements_by_xpath("//div[@class='specfpro_details_cont']/div/div/a")
for i in NM: #Scraping title data
    nm.append(i.text)
PR = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[3]")
for i in PR:#Scraping processor data
    pr.append(i.text.split(': ')[1])
DS = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[4]")
for i in DS:#Scraping display data
    ds.append(i.text.split(': ')[1])
OS = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[5]")
for i in OS:#Scraping os data
    os.append(i.text.split(': ')[1])
ME = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[6]")
for i in ME:#Scraping memory data
    me.append(i.text.split(': ')[1])
GR = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[7]")
for i in GR:#Scraping graphics data
    gr.append(i.text.split(': ')[1])
BD = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[8]")
for i in BD:#Scraping bosy weightand size data
    bd.append(i.text.split(': ')[1])
UR = driver.find_elements_by_xpath("//div[@class='specfpro_details_cont']/div/div/a")
for i in UR:#Scraping url data
    ur.append(i.get_attribute('href'))
data = pd.DataFrame({'Title':nm, 'Processor':pr, 'Display':ds, 'OS':os, 'Memory':me, 
                     'Graphics':gr, 'Body size/weight':bd, 'URL':ur }) #Creating a dataframe
driver.close() #closing the browser
data #Displaying the dataframe
```

8. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
#For scraping 1st page data
driver = webdriver.Chrome("chromedriver.exe") #calling webdriver
driver.get('https://www.forbes.com')  #getting into website
time.sleep(5)
driver.maximize_window() #Maximizing the window
driver.find_element_by_xpath("//div[@class='header__left']").click() #clicking left header
time.sleep(5)
driver.find_element_by_xpath("//div[@class='mpBfVZz3']").click() #clicking billionaires
driver.find_element_by_xpath("//li[@class='TjJgrPSg _2bNo56RE secondary']").click() #clicking world billionaires
time.sleep(5)
rk = []
nm = []
nw = []
ag = []
ci = []
so = []
ir = []
RK = driver.find_elements_by_xpath("//div[@class='rank']")
for i in RK: #Scraping rank data
    rk.append(i.text)
NM = driver.find_elements_by_xpath("//div[@class='personName']")
for i in NM: #Scraping name data
    nm.append(i.text)
NW = driver.find_elements_by_xpath("//div[@class='netWorth']")
for i in NW: #Scraping net worth data
    nw.append(i.text)
AG = driver.find_elements_by_xpath("//div[@class='age']")
for i in AG: #Scraping age data
    ag.append(i.text)
CI = driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
for i in CI: #Scraping country data
    ci.append(i.text)
SO = driver.find_elements_by_xpath("//div[@class='source']")
for i in SO: #Scraping source data
    so.append(i.text)
IR = driver.find_elements_by_xpath("//div[@class='category']")
for i in IR: #Scraping industry data
    ir.append(i.text)
data = pd.DataFrame({'Rank':rk, 'Name':nm, 'Age':ag, 'Country':ci, 'Source':so,'Industry':ir}) #Creating a dataframe
print('1st page data scraped successfully')
data.head()
```

9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver = webdriver.Chrome("chromedriver.exe") #calling webdriver
driver.get('https://www.youtube.com/watch?v=g6fnFALEseI') #getting into youtube url / video link
driver.maximize_window() #Maximizing the window
for _ in range(400): #Scroll down to generate more comments
    driver.execute_script("window.scrollBy(0,400)")
    time.sleep(0.1) #Giving browser time to load page and scrape dynamically
cm = []
up = []
tm = []
CM = driver.find_elements_by_xpath("//div[@id='comment-content']/ytd-expander/div/yt-formatted-string")
for i in CM: #Scraping comment content data
    cm.append(i.text)
UP = driver.find_elements_by_xpath("//*[contains(@aria-label,'Like this comment along with ')]")
for i in UP: #Scraping number of upvotes data
    up.append(i.get_attribute('aria-label').split('with ')[1].split(' other')[0])
TM = driver.find_elements_by_xpath("//div[@class='style-scope ytd-comment-renderer']/yt-formatted-string/a")
for i in TM: #Scraping time of comment data
    tm.append(i.text)
data = pd.DataFrame({'Comment':cm[0:500], 'No. of upvotes':up[0:500], 'Time of comment':tm[0:500]}) #Saving into data frame
print(data.shape) #Checing size of data
data #displaying dataframe
```
